# Exploratory Data Analysis

```python
dsale_df.head()
```

```python
# Usind bar chart for gender count
genplot=sn.countplot(x='Gender',data=dsale_df,hue='Gender')
for bar in genplot.containers:
    genplot.bar_label(bar)
plt.show()
```

Above graph showing gender count

```python
#plot a chart for Amount vs Gender
plt.figure(figsize=(10,5))
sale_gen=dsale_df.groupby('Gender').agg(Amount=('Amount', 'sum'))
genplot=sn.barplot(x='Gender',y='Amount',data=sale_gen,hue='Gender')
for bar in genplot.containers:
    genplot.bar_label(bar)
plt.show()
```

From above graph we can say that femals purchasing power is more than men and most of buyers are femalse

```python
count=sn.countplot(data=dsale_df,x='Age Group' ,hue='Gender')
for bar in count.containers:
    count.bar_label(bar)
plt.show()
```

From this graph we can say that most of buyers are from 26-35 age group

```python
Ag_Amt=dsale_df.groupby('Age Group').agg(Amount=('Amount','sum'))
```

```python
Ag_Amt
```

```python
Ag_Amt=dsale_df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

agebar=sn.barplot(x='Age Group',y='Amount',data=Ag_Amt,hue='Age Group')
```

This Graph is showing that most of the buyers are from 26-35 age group

```python
# Total no of orders for top 10 state 
plt.figure(figsize=(15,5))
ord=dsale_df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders", ascending=False).head(10)
sn.barplot(data=ord,x="State",y="Orders",hue="State")
plt.show()
```

```python
# Total sales  for top 10 state 
plt.figure(figsize=(15,5))
ord=dsale_df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sn.barplot(data=ord,x="State",y="Amount",hue="State")
plt.show()
```

From above graph we  ca see that the Uttar pradesh state have most no of orders & amount/sales

# Marital Status

```python
mrt=sn.countplot(data=dsale_df,x='Marital_Status',hue='Marital_Status')

for bar in mrt.containers:
    mrt.bar_label(bar)
plt.show()
```

```python
mrg=dsale_df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum()
sn.barplot(data=mrg,x='Marital_Status',y='Amount',hue='Gender')
plt.show()
```

 From this graph we can see that most of buyers are married women

# Occupation

```python
plt.figure(figsize=(20,5))
cnt=sn.countplot(x='Occupation',data=dsale_df)
for bar in cnt.containers:
    cnt.bar_label(bar)
```

```python
plt.figure(figsize=(20,5))
occ=dsale_df.groupby(['Occupation'], as_index=False)['Amount'].sum()
sn.barplot(data=occ,x='Occupation',y='Amount')
plt.show()
```

# Product Category

```python
dsale_df
```

```python
plt.figure(figsize=(25,5))
prd=sn.countplot(x='Prod_Categ',data=dsale_df)
for bars in prd.containers:
    prd.bar_label(bars)
plt.show()
```

```python
prodplot = dsale_df.groupby(['Prod_Categ'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20,5))
sn.barplot(data = prodplot, x = 'Prod_Categ',y= 'Amount')
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
pplot = dsale_df.groupby(['Prod_Categ'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(20,5))
sn.barplot(data = pplot, x = 'Prod_Categ',y= 'Orders')
plt.show()
```
